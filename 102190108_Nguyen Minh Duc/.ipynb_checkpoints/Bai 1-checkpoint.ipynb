```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data1=np.random.default_rng().normal(100,20,1000)
```

```python
data2=np.random.default_rng().normal(50,10,1000)
```

```python
data3=data2-data1
```

```python
print(f"Do lech chuan mau cua data 1: {data1.std()}")
```

```python
print(f"Do lech chuan mau cua data 3: {data3.std()}")
```

```python
print(f"Gia tri trung binh mau cua data 1: {data1.mean()}")
```

```python
print(f"Gia tri trung binh mau cua data 3: {data3.mean()}")
```

```python
fig=plt.figure(figsize=(12,6))
ax=fig.add_axes([0,0,1,1])
ax.plot(data1,data3,"ro")
plt.xlabel("data1")
plt.ylabel("data3")
```

```python
data={"Data 1":data1,"Data 3":data3}
df=pd.DataFrame(data)
```

```python
print("He so tuong quan cua data 1 va data 3")
df.corr()
```
